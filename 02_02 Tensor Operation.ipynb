```python
import torch
import numpy
```

```python
# create 2 tensors

u = torch.tensor([1.0, 2.0])
v = torch.tensor([3.0, 4.0])

print(u)
print(v)
```

```python
# addition and subtraction
print(u + v)
print(u - v)
```

```python
# generate a torch tensor with 2x4 matrix using random function (distribution of 0-1)
a = torch.rand(2, 4) * 2 - 1
# common operations: abs, ceil, floor, clamp
print('Common functions:')
print(torch.abs(a))
print(torch.ceil(a))
print(torch.floor(a))
print(torch.clamp(a, -0.5, 0.5))
```

```python
# reduction -> mean

x = torch.tensor([1, 2, 3, 4, 5], dtype=torch.float32)
print(x)
print(x.mean())
```

```python
# max and min

print(x.max())
print(x.min())
```

```python
# trigonometric functions and their inverses

import numpy as np

x = torch.tensor([0, np.pi / 2, np.pi])
print(x)
print(torch.sin(x))
print(torch.cos(x))
```

```python
pi = torch.linspace(-np.pi/2, np.pi/2, steps=1000)
print(pi[:5])  # lower bound
print(pi[-5:]) # upper bound

sined = torch.sin(pi)
cosed = torch.cos(pi)
print(sined[0:5])
print(cosed[0:5])
```

```python
# get an evenly spaced list of numbers between a range
pi = torch.linspace(-np.pi/2, np.pi/2, steps=1000)
print(pi[:5])  # lower bound
print(pi[-5:]) # upper bound

sined = torch.sin(pi)
cosed = torch.cos(pi)
print(sined[0:5])
print(cosed[0:5])
```

```python
# plot

import matplotlib.pyplot as plt

plt.plot(sined, label="sined")
plt.plot(cosed, label="cosed")
plt.legend()
plt.show()
```
